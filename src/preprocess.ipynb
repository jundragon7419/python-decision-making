
### 데이터 타입과 결측치 확인

```python
import pandas as pd
```

```python
file_paths = [
    "../data/raw/2015-2018.xlsx",
    "../data/raw/2019.xlsx",
    "../data/raw/2020.xlsx",
    "../data/raw/2021.xlsx",
    "../data/raw/2022.xlsx",
    "../data/raw/2023.xlsx",
    "../data/raw/2024.xlsx",
]
```

```python
def check_file(path: str) -> None:
    """파일 하나(모든 시트)를 읽어 dtype과 결측치를 출력한다. 가공하지 않음."""
    sheet_dict = pd.read_excel(path, sheet_name=None, engine="calamine") # 기존 엔진은 너무 느려 Rust를 사용한 엔진 사용
    df = pd.concat(sheet_dict.values(), ignore_index=True) # 시트로 나누어진 자료들을 하나로 모음
 
    print(f"\n{'='*60}")
    print(f"파일: {path}")
    print(f"{'='*60}")
    print(f"전체 행 수: {len(df)}")
    print(f"\ndtype:\n{df.dtypes}")
    print(f"\n결측치:\n{df.isna().sum()}")
```
